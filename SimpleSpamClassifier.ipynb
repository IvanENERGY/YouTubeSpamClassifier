```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer # extract text->vector
from sklearn.linear_model import LogisticRegression #use for 2 class classification
from sklearn.metrics import accuracy_score

```

```python
df= pd.read_csv('mail_data.csv')
```

```python
df
```

```python
df.isnull().sum()
```

```python
data = df.where(pd.notnull(df),'')
```

```python
df.info()
```

```python
data.shape
```

```python
data.loc[data['Category']=='spam','Category',]=0
data.loc[data['Category']=='ham','Category',]=1
```

```python
data.head()
```

```python
X=data['Message']
Y=data['Category']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train,X_test,Y_train,Y_test= train_test_split(X,Y,test_size=0.2,random_state=3)
#randomstate: Controls the shuffling applied to the data before applying the split. Pass an int for reproducible output across multiple function calls
```

```python
print(X.shape)
print(X_train.shape)
print(X_test.shape)
```

```python
print(Y.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
#Convert a collection of raw documents to a matrix of TF-IDF features.
#mindf= ignore terms that have a document frequency strictly lower than the given threshold
# ‘english’ is currently the only supported string value
#lowercase:Convert all characters to lowercase before tokenizing.
feature_extraction=TfidfVectorizer(min_df=1,stop_words='english',lowercase=True)
X_train_features= feature_extraction.fit_transform(X_train)
X_test_features=feature_extraction.transform(X_test)

Y_train=Y_train.astype('int')
Y_test=Y_test.astype('int')
```

```python
print(X_train)
```

```python
print(X_train_features)
```

```python
model = LogisticRegression()
model.fit(X_train_features,Y_train)
```

```python
prediction_on_training=model.predict(X_train_features)
score=accuracy_score(Y_train,prediction_on_training)
print('accuracy on training data %s'%score)
```

```python
prediction_on_test=model.predict(X_test_features)
score=accuracy_score(Y_test,prediction_on_test)
print('accuracy on test data %s'%score)
```

```python
input_your_mail=["This is the 2nd time we have tried to contact u. U have won the prizr. 2 claim is easy, just call 52552"]
input_data_features=feature_extraction.transform(input_your_mail)
prediction=model.predict(input_data_features)
print(prediction) #spam=0

if prediction[0]==0:
    print('spam')
else:
    print('ham')
```
